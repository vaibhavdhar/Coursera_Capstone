```python
import pandas as pd
listofhoods = pd.read_html("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M", header=0)
df = pd.DataFrame(listofhoods[0]) 
df = df[df.Borough != 'Not assigned']
df.loc[(df['Neighbourhood'] ==  'Not assigned'), 'Neighbourhood'] = df['Borough']
df = df.groupby(['Postcode','Borough']).agg(', '.join).reset_index(drop = False)
df
```

## Adding the latitude and longitude data to the dataframe

```python
longlat = pd.read_csv('https://cocl.us/Geospatial_data')
geodf = pd.DataFrame(longlat)
df['Latitude'] = geodf['Latitude']
df['Longitude'] = geodf['Longitude']
df
```

```python
from geopy.geocoders import Nominatim
import folium

address = 'Toronto, ON'
geolocator = Nominatim(user_agent="tor_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))



# create map of New York using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)

# add markers to map
for lat, lng, Borough, Neighbourhood in zip(df['Latitude'], df['Longitude'], df['Borough'], df['Neighbourhood']):
    label = '{}, {}'.format(Neighbourhood, Borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto)  
    
map_toronto
```
